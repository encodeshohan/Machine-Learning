**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Data**

image.png

```python
height = np.array([167,182,176,173,172,174,169,173,170])
weight = np.array([51,62,69,64,65,56,58,57,55])
class_label = np.array(['Underweight','Normal','Normal','Normal','Normal','Underweight','Normal','Normal','Normal'])
```

**Plot the data using matplotlib**

```python
# prompt: plot X1 and X2 as points as red for Underweight and green for normal

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(height[class_label == 'Underweight'], weight[class_label == 'Underweight'], color='red', label='Underweight')
plt.scatter(height[class_label == 'Normal'], weight[class_label == 'Normal'], color='green', label='Normal')
plt.scatter(170, 57, color='black', label='unknown?')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Scatter Plot of height and weight')
plt.legend()
plt.show()
```

**Implementing the Eucledian Distance Method**
<br> between two data points (x1, y1) and (x2,y2)

image.png

```python
# Euclidean distance calculation
def euclidean_distance(point1, point2):
  return np.sqrt(np.sum((point1 - point2)**2)).round(2)

# Given point
point = np.array([170, 57])

# Calculate distances
distances = []
for i in range(len(height)):
  distances.append(euclidean_distance(point, np.array([height[i], weight[i]])))

distances
```

```python
# prompt: order values in distances and take first 3 to cout max class
K = 3
# Create a DataFrame for better handling
df_distances = pd.DataFrame({'distance': distances, 'class': class_label})

# Sort by distance and get the first 3
nearest_neighbors = df_distances.sort_values(by='distance').head(K)

# Count the occurrences of each class among the nearest neighbors
class_counts = nearest_neighbors['class'].value_counts()

# Get the class with the maximum count
max_class = class_counts.index[0]

print(f"The majority class among the {K} nearest neighbors is: {max_class}")
```
